#### Q3: Demonstrate remove_sparse_columns function in a Jupyter Notebook by:
* removing columns representing parties that received fewer votes than a specified threshold.

```python
threshold = 1000
filtred_df = ex1.remove_sparse_columns(kneset25,threshold)

print(f"before filter, df shape was:{kneset25.shape}, after filter, shape is:{filtred_df.shape}", end='\n\n')
print(f"{kneset25.shape[1]-filtred_df.shape[1]} columns were removed:")
for col in kneset25.columns.tolist():
    if col not in filtred_df.columns.tolist():
        print(f"\t{col}", end="")
        print(" "*(30-len(col)),end="")
        print(f"\t\t {kneset25[col].sum()}")
              
```
